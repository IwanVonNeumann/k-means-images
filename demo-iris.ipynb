## Iris flowers k-means clustering

```python
import os

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
DATA_DIR = 'data'
IRIS_FILE_NAME = 'iris.data.csv'
```

```python
iris_file_path = os.path.join(DATA_DIR, IRIS_FILE_NAME)
```

```python
df = pd.read_csv(iris_file_path)
```

```python
df.shape
```

```python
df.head()
```

```python
X = df.values[:, [0,2]]
y = df.values[:, -1]
```

```python
class_color_map = {
    'Iris-setosa': 'blue',
    'Iris-versicolor': 'red',
    'Iris-virginica': 'lime'
}
```

```python
class_colors = [class_color_map[label] for label in y]
```

```python
cluster_color_map = {
    0: 'blue',
    1: 'red',
    2: 'lime'
}
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

print('done in {} steps'.format(kmeans.n_iter_))
```

```python
cluster_colors = [cluster_color_map[label] for label in kmeans.labels_]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
ax1.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=class_colors, edgecolors='black')
ax1.grid()
ax1.set_title('True labels')
ax1.set_xlabel("sepal length")
ax1.set_ylabel("petal length")

ax2.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=cluster_colors, edgecolors='black')
ax2.grid()
ax2.set_title('K-means labels')
ax2.set_xlabel("sepal length")
ax2.set_ylabel("petal length")

plt.show()
```
