```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)
```

```python
def scatter(X, colors='white'):
    plt.figure(figsize=(6, 4))
    plt.scatter(X[:, 0], X[:, 1], marker='o', s=50, c=colors, edgecolors='black')
    plt.grid()
    plt.show()
```

```python
scatter(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans
print('done in {} steps'.format(kmeans.n_iter_))
```

```python
color_map = {
    0: 'red',
    1: 'green',
    2: 'blue'
}
```

```python
colors = [color_map[y] for y in kmeans.labels_]
```

```python
scatter(X, colors=colors)
```
